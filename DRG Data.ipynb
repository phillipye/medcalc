# Collecting Medicare DRG Payment Data

Using data from: https://data.cms.gov/Medicare-Inpatient/National-Summary-of-Inpatient-Charge-Data-by-Medic/efwk-h4x3

```python
import pandas as pd
```

```python
raw_data = pd.read_csv("drg_data.csv")
raw_data.columns = [x.strip() for x in raw_data.columns]
raw_data.head(5)
```

```python
clean_data = raw_data.copy(deep=True)
clean_data["DRG Code"] = [x.split(" -")[0] for x in clean_data["DRG Definition"]]
clean_data["DRG Definition"] = [x.split("- ")[1] for x in clean_data["DRG Definition"]]
output_data = clean_data[["DRG Code", "DRG Definition", "Total Discharges", "Average Covered Charges", "Average Total Payments", "Average Medicare Payments"]]
output_data.index =  output_data["DRG Definition"] + " - " + output_data["DRG Code"]
output_data[["Average Covered Charges", "Average Total Payments", "Average Medicare Payments"]] = output_data[["Average Covered Charges", "Average Total Payments", "Average Medicare Payments"]].applymap(lambda x : float(x.replace("$", "")))
output_data.to_json("drg_avg_costs.json", orient="index")
output_data.head(5)
```
